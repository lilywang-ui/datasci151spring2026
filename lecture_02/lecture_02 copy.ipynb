# Summary of Main Code Lines in Lecture 2


## 1. **Importing Libraries**


```python
import math
import statistics as stats
import matplotlib.pyplot as plt
import pandas as pd
```



## 2. **Loading Data**


```python
carfeatures = pd.read_csv("data/features.csv")
```



## 3. **Displaying Data**


```python
display(carfeatures)  # All data
display(carfeatures['cylinders'])  # Single column
```



## 4. **Creating Frequency Tables**


```python
table = pd.crosstab(index=carfeatures['cylinders'], columns="count")
display(table)
```



## 5. **Summary Statistics**


```python
display(carfeatures.describe())  # Basic stats
display(round(carfeatures.describe(), 2))  # Rounded to 2 decimals
display(carfeatures.describe(percentiles=[0.33, 0.66]))  # Custom percentiles
```



## 6. **Creating Visualizations**


```python
plt.scatter(x=carfeatures['weight'], y=carfeatures['mpg'])
plt.show()
```
