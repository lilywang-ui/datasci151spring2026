1. Objects and Data Types
Almost everything in Python is an object with a corresponding type (or class)
Common data types include:
Integers (int): whole numbers
Floats (float): numbers with decimals
Strings (str): text sequences
Use type() function to check an object's type
Use print() to display multiple outputs in one cell
2. Variables
Variables are named storage locations that hold values
Created using the assignment operator = (not an equals sign)
Naming rules:
Can include letters, numbers, and underscores
Cannot start with a number
Are case-sensitive
Should be descriptive for readability
Variables can hold any data type (numbers, strings, DataFrames, etc.)
3. Lists
Collections of multiple objects stored in a single named location
Created using square brackets [...] with elements separated by commas
Key features:
Use len() to count elements
Zero-indexed: first element is at position 0
Access elements using square bracket notation: list[index]
Can contain mixed data types
Can contain other lists as elements
Elements can be modified using assignment
